```python
def sql_to_df(query):
    return pd.read_sql(query,con)
```

```python
query1= '''SELECT first_name, last_name
                    FROM customer; '''
sql_to_df(query1).head()
```

#### Aggregate function
+ AVG() - Returns the average value.
+ COUNT() - Returns the number of rows.
- FIRST() - Returns the first value.
+ LAST() - Returns the last value.
+ MAX() - Returns the largest value.
+ MIN() - Returns the smallest value.
* SUM() - Returns the sum.

```python
query11= '''SELECT COUNT(customer_id)
                        FROM customer;'''
sql_to_df(query11)
```

#### Wildcards
| wildcard | description       |
|----------|-------------------|
| %        | zero or more char |
| _        | single char       |
| [ ]      | specific char's   |

```python
query12='''SELECT *
                FROM customer
                WHERE first_name LIKE 'M%'     '''   # any customers whose first names start with 'M'
sql_to_df(query12).head()
```

```python
query13='''SELECT *
                    FROM customer 
                    WHERE last_name LIKE '_ING'   '''
sql_to_df(query13)
```

#### IMPORTANT NOTE!
Using [charlist] with SQLite is a little different than with other SQL formats, such as MySQL.  
In MySQL you would use:  
WHERE value **LIKE '[charlist]%'**  
In SQLite you use:  
WHERE value __GLOB '[charlist]*'__

```python
query14='''SELECT * 
                    FROM customer 
                    WHERE first_name GLOB '[AB]*' ;  '''  # Select any customers whose first name begins with an A or a B
sql_to_df(query14).head()
```

#### ORDER BY
ascending order by default. Add DESC for descending order

```python
query15= '''SELECT *
                    FROM customer
                    ORDER BY last_name; '''
sql_to_df(query15)
```

```python
query16= '''SELECT *
                    FROM customer
                    ORDER BY last_name DESC; '''
sql_to_df(query16)
```

#### GROUP BY
used w/ aggregate fn
SELECT column_name, aggregate_function(column_name)   
FROM table_name   
WHERE column_name operator value   
GROUP BY column_name;  

```python
query17= '''SELECT store_id, COUNT(customer_id)
                        FROM customer
                        GROUP BY store_id; '''
sql_to_df(query17)
```
