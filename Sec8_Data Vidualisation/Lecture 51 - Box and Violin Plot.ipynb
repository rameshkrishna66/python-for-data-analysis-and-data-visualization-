```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
url = 'http://en.wikipedia.org/wiki/Box_plot#mediaviewer/File:Boxplot_vs_PDF.svg'
```

```python
data1=randn(100)
data2=randn(100)
```

```python
sns.boxplot(y=[data1,data2])  # supposed to output both data1 and data2 in the same box, but it doesnt...
```

```python
fig,(axis1,axis2)=plt.subplots(ncols=2,sharey=True)
sns.boxplot(y=data1,ax=axis1)
sns.boxplot(y=data2,ax=axis2)
```

```python
fig,(a1,a2) = plt.subplots(ncols=2,sharey=True)
sns.boxplot(y=data1,whis=np.inf,ax=a1)  # include the outlier denoted as dot
sns.boxplot(y=data2,whis=np.inf, ax=a2)
```

```python
# example of a box plot not showing the full picture 

# Normal Dist
data1= stats.norm(loc=0,scale=5).rvs(100)  
# `scipy.nonr` for a normal continuas random variable w/ `loc` and `scale` representing mean and var, res. 
#  `rvs()` for random variates of given type. len(data1)=100

# two gamma dist. Concatenated together 
data2= np.concatenate([stats.gamma(5).rvs(50)-1,
                                           -1*stats.gamma(5).rvs(50)])
# len(data2)=100

# box plot both data1 and data2
fig,(n,g) =plt.subplots(ncols=2, sharey=True)
sns.boxplot(y=data1,whis=np.inf, ax=n) 
sns.boxplot(y=data2,whis=np.inf, ax=g)
```

```python
# both boxplots looks similar but they actually are very diff ->violin plot
fig,(n,g) = plt.subplots(ncols=2,sharey=True)
sns.violinplot(y=data1, ax=n)
sns.violinplot(y=data2, ax=g)
```

```python
# We can also change the bandwidth of the kernel used for the density fit of the violin plots if desired
sns.violinplot(y=data2,bw=0.01)
```

```python
# include individual points (like rug plot)
sns.violinplot(y=data1,inner="stick")
```
