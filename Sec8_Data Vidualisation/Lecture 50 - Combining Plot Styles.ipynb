```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dataset= randn(100)
```

```python
sns.distplot(dataset,bins=25)  # combines histogram with kde by default
```

```python
sns.distplot(dataset,bins=25,rug=True,hist=False) # kde and rug
```

```python
# pass dictionary
sns.distplot(dataset,bins=25,
                     kde_kws={"color":"indianred","label":"KED PLOT"},
                     hist_kws={"color":"blue","label":"HIST"})
```

```python
ser1= pd.Series(dataset, name="My_Data")
```

```python
sns.distplot(ser1,bins=25)
```
