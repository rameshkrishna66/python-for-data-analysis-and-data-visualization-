```python
import numpy as np
from numpy.random import randn
import pandas as pd
```

```python
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
dataset1=randn(100)
```

```python
plt.hist(dataset1)   # bins=10 by default
```

```python
dataset2=randn(80)
plt.hist(dataset2, color="indianred")
```

```python
# two histograms in the same graph
plt.hist(dataset1,normed=True,color="indianred",alpha=0.5,bins=20)# `normed=T` because the size of the dataset is diff. `alpha` controls the transparency
plt.hist(dataset2,normed=True,alpha=0.5,bins=20)
```
